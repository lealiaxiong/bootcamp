```python
import numpy as np
import pandas as pd
import scipy.special

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('data/retina_spikes.csv', comment='#')
df.head()
```

```python
p = bokeh.plotting.figure(
    frame_height=150,
    frame_width=600,
    toolbar_location='above',
    x_axis_label='t (ms)',
    y_axis_label='V (uV)',
)

p.line(
    source=df,
    x='t (ms)',
    y='V (uV)',
    line_width=2,
)

bokeh.io.show(p)
```

```python
x = np.linspace(-15, 15, 400)

norm_I = 4 * (scipy.special.j1(x) / x)**2
```

```python
p = bokeh.plotting.figure(
    frame_height=250,
    frame_width=550,
    x_axis_label='x',
    y_axis_label='norm. intensity'
)

p.line(
    x=x,
    y=norm_I,
    line_width=2,
)

bokeh.io.show(p)
```
