```python
import tqdm

import numpy as np
import bokeh.plotting
import bokeh.io

bokeh.io.output_notebook()
```

```python
# This cell takes 3.5 hrs to run!

rg = np.random.default_rng()

x = [1, 10, 100, 1000, 10000, 100000]
y = np.zeros(len(x))

for j, n in tqdm.tqdm(enumerate(x)):
    data = np.arange(n)
    
    # Number of bootstrap samples
    n_bs = 2000

    frac_not_ins = np.zeros(n_bs)

    for i in tqdm.tqdm(range(n_bs)):
        bs_sample = rg.choice(data, size=n, replace=True)
        not_in = [val for val in data if val not in bs_sample]
        frac_not_ins[i] = len(not_in) / n
        
    y[j] = np.mean(frac_not_ins)
```

```python
p = bokeh.plotting.figure(
    frame_width=400,
    frame_height=300,
    x_axis_label='Number of measurements',
    y_axis_label='Average fraction of measurements that do not appear in a bootstrap sample',
    x_axis_type='log',
)

p.circle(
    x,
    y,
)

# Theoretical

x_theor = np.logspace(0, 5, 400)
y_theor = (1 - 1/x_theor)**x_theor

p.line(
    x_theor,
    y_theor,
    color='orange'
)

bokeh.io.show(p)
```
