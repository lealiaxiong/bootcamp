# Boolean indexing

```python
df.loc[df['participant number']==27, 'percent correct']
```

```python
df['participant number'] == 27
```

```python
df.loc[df['gender']=='f', 'percent correct']
```

```python
df.loc[df['age']>50, 'percent correct']
```

```python
df.loc[df['age']>50, :]
```

```python
df.loc[(df['age'] < 21) & (df['gender'] == 'f'), :]
```

The `&` is a bitwise operator, so you need to put both criteria in parentheses.

For bitwise OR, we use `|`.

```python
inds = (df['age'] < 30) & (df['gender'] == 'f') & (df['percent correct'] > 85)

df.loc[inds, :]
```

An insomniac is a person with Sleep Condition Indicator <= 16. Let's calculate if each participant is an insomniac.

```python
df['sci'] <= 16
```

Add it to the dataframe:

```python
df['insomnia'] = df['sci'] <= 16

df.head()
```

```python
df.loc[df['insomnia'], 'percent correct']
```

```python
# Mean percent correct for people with insomnia
np.mean(df.loc[df['insomnia'], 'percent correct'])
```

```python
# Mean percent correct for people without insomnia
np.mean(df.loc[~df['insomnia'],'percent correct'])
```

```python
# Summary statistics for the whole dataframe
df.describe()
```

```python
# Output data to .csv file
df.to_csv('data/gfmt_sleep_with_insomnia.csv', index=False)
```

## Split-apply-combine

```python
# Split
grouped = df.groupby('insomnia')
```

```python
# Apply (mean function) and combine
grouped.mean()
```

Note that now the indices are `False` and `True` and the indices have the name `insomnia`.

```python
# Reset the index
grouped.mean().reset_index()
```

```python
grouped['percent correct'].mean().reset_index()
```

```python
# Method chaining
df.groupby('insomnia')['percent correct'].mean().reset_index()
```

```python
df['within insomnia group rank'] = grouped['percent correct'].rank()
df.head()
```

```python
df_sorted = df.sort_values(by=['insomnia', 'percent correct'])

df_sorted[['insomnia', 'percent correct', 'within insomnia group rank']]
```

```python
def coeff_of_var(data):
    """Compute coefficient of variation for an array of data."""""
    return np.std(data) / np.mean(data)
```

```python
df.groupby('insomnia').agg(coeff_of_var).reset_index()
```

```python
for name, group in df.groupby(['gender', 'insomnia']):
    print(name, ':', group['percent correct'].mean())
```
